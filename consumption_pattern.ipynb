```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
pd.set_option('display.max_rows',None)
```

```python
data = pd.read_excel("consumption.xlsx")
data
```

```python
data =data.drop(["Unnamed: 1","Unnamed: 2","Unnamed: 3","Unnamed: 4","Unnamed: 5","Unnamed: 6","Unnamed: 7","Unnamed: 8","Unnamed: 9"],axis = 1)
```

```python
data.columns
```

```python
data = data.drop("Unnamed: 13", axis=1)
```

```python
data.columns
```

```python
data["Item Description"] = data["Item Description"].str.lower()
data["Item Description"]
```

## Total Consumptions Pattern of Last 6 Months

```python
total_cons = data[["Item Description","Consumption Quantity"]].sort_values("Consumption Quantity",ascending = False)
total_cons
```

```python
g=sns.relplot(data=total_cons,x= "Item Description",y = "Consumption Quantity")
g.fig.set_figheight(6)
g.fig.set_figwidth(30)
plt.xticks(rotation=90)
plt.ylabel([])
plt.title( "Consumption Pattern", weight = "bold",size = 15)

plt.show()
```

## Top 10 Consumable Items in Last 6 Months

```python
top_10 = total_cons.head(11)
top_10
```

```python
plt.barh(top_10["Item Description"],top_10["Consumption Quantity"],color = "#f24162" , alpha = 0.4)
plt.xticks([])
plt.title( "TOP 10 CONSUMABLE ITEMS", weight = "bold",size = 15)
plt.show()
```

## ABC ANALYSIS ON THE BASIS OF CONSUMPTION

```python
data.head()
```

```python
cs = data["Avg. Cons Per Month"]
```

```python
print(cs)
```

```python
category = []

for i in cs:
    if i >= 10000:
        category.append("A")
    elif  5000 <= i < 10000:
        category.append("B")
    else:
        category.append("C")
        
    
```

```python
category
```

```python
data["Category"]= category
```

```python
data.sort_values("Category")
```

```python
category = data["Category"].value_counts()
category
```
